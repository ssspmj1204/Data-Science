# Numpy란? 
### 과학 계산을 위한 라이브러리로서 다차원 배열을 처리하는데 필요한 여러 유용한 기능 제공
Numpy 라이브러리 안에는 다차원배열을 지원하는 클래스 존재 -> **numpy.ndarray**
- 배열은 동일한 타입값을 가짐
- 다(N) ~ 차원 형태로 구성
- 배열의 차원  : rank
- 각 차원의 크기 : shape

* Numpy 패키지 설치 명령어 : **pip install numpy**

```
import numpy as np
```

##  1. 배열 생성


*   배열 생성 함수 : np.array()
*   배열 생성 유용 함수 : np.arange(start, stop)
*   배열 차원 변경 함수 : np.ndarray.reshape(행, 열)

```
arr1 = np.array([1,2,3,4,5]) 
arr2 = np.array([[1,2,3],[4,5,6]])

# 0이상 100미만까지 10씩 차이나게 생성 
print(np.arange(0,100 ,10))
# 0이상 1미만까지 0.1씩 차이나게 생성
print(np.arange(0,1,0.1))
```

```
# reshape() -> 배열 크기 설정
arg_arr = np.arange(1,51) 
arg_arr = arg_arr.reshape(5,10)
print(arg_arr)
print()
print(np.arange(0,100 ,10).reshape(2,5))
print()
print(np.arange(0,1,0.1).reshape(5,2))
```

```
# 모든 원소를 1차원으로 변경
print(arg_arr.ravel())
```


* 0으로 구성된 N차원 배열 생성 : np.zeros(shape)
* 1로 구성된 N차원 배열 생성 : np.ones(shape)
* 초기화 되지 않은 N차원 배열 생성 : np.empty(shape)


```
arr_zero = np.zeros((2,4))
print("zeros: ", arr_zero)

arr_one = np.ones((3,2))
print("ones: ", arr_one)

arr_empty = np.empty((2,2))
print("empty: ", arr_empty)
```




## 2. 배열 속성





```
# 배열의 크기
print(arr1.shape)
print(arr2.shape)
```

```
# 배열 요소 개수(배열 사이즈)
print(arr1.size)
print(arr2.size)
```

```
# 배열 데이터 타입 확인
print(arr1.dtype)
print(arr2.dtype)
```

```
# 배열 차원(rank) 확인
print(arr1.ndim)
print(arr2.ndim)
```

## 3. 데이터 접근하기



###  슬라이싱



```
arr = np.arange(1,51).reshape(5,10)
arr.ndim
```

```
# 1번째 행까지와 모든 열에 접근
print(arr[0:2])
print()
print(arr[0:2,:])
```

```
# 1번째 행까지와 7번째열부터 끝열까지 접근
print(arr[0:2, 7:])
```

```
print(arr[2:,2:7])
```


### 정수 인덱싱

```
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

```
print(arr2[0])
print(arr2[1]) 
```

```
print(arr2[0][0])
print(arr2[1][2])
print(arr2[1][0])
```

```
print(arr2[1,2])
print(arr2[0,2])
```

```
print(arr2[[0,1,1],[1,0,2]])
```

### 불리언 인덱싱


```
arr2 = np.array([[1,2,3,4,5],[55,0,60,80,90]])
arr2
```

```
print(arr2[1])
print(arr2[1] != 0)
```

```
# 0이 아닌 요솟값에 접근
arr2[1][arr2[1] != 0]
```

```
# 짝수 요솟값에 접근
arr2[0:][arr2[0:] % 2 == 0]
```

```
arr2[0:][arr2[0:] >= 70]
```

## 3. 배열 연산

```
arr3 = np.array([[1,2],[3,4]])
arr4 = np.array([[3,4],[5,6]])
```

### 기본연산

```
print(arr3 + 3)
```

```
print(arr3/2.0)
```

```
print(arr3+arr4)
```

```
print(arr3*arr4 )
```

```
np.array([2,3,4]+[3,5,7])
```

###  행렬 연산

```
np.dot(arr3,arr4)
```

```
print(arr3 @ arr4)
```

## 4. 배열 관련 유용한 함수

```
print("arr3:", arr3)
print("arr4:", arr4)
```

```
arr5 = np.array([[3,1,2,7],[9,7,2,0]])
print(arr5.max())
print(arr5.min())
print(arr5.argmax())  # 모든 요소 중 최댓값의 인덱스
print(arr5.argmin())  # 모든 요소 중 최소값의 인덱스
print(arr5.cumsum())  # 누적합
```

```
for i in arr5:
  print(i)
```

```
for i in arr5.flat:
  print(i, end = ' ')
```

```
# 합
print(np.sum(arr3))
print(arr4.sum())
```

```
# 제곱근
print(np.sqrt(arr3))
```

```
# 평균
np.mean(arr3)
```
