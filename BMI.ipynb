# 많은 인원의 체질량지수(BMI) 계산

```python
import numpy as np
```

#### 0. 데이터 불러오기

```python
data = np.loadtxt("height_weight.txt", delimiter=',')
data
```

#### 1. 데이터 속성 확인
- 배열 크기
- 배열 차원수
- 배열 전체 요소 개수

```python
data.shape
```

```python
data.ndim
```

```python
data.size
```

#### 2. 배열 데이터에서 키와 몸무게 데이터 각각 변수에 넣기

```python
height = data[0] * 0.01
height
```

```python
weight = data[1]
weight
```

#### 3. BMI지수 공식에 대입해보기

```python
bmi = weight / (height ** 2)
bmi
```

#### 4. 과체중 이상인 사람은 몇명인지 구하기

```python
# bmi >= 23
bmi[bmi >= 23]
```

```python
len(bmi[bmi >= 23])
```
