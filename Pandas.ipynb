### Pandas 란?
* 데이터 처리와 분석에 최적화된 라이브러리
* 행과 열로 이루어진 데이터 객체를 만들어서 다룰 수 있음
* 빅데이터를 처리하는데 매우 편리한 함수들을 제공

```
import pandas as pd
```

#### 1. Pandas 구조
* Series : 1차원 배열 데이터
* DataFrame : 2차원 표 형태의 배열 데이터

* 컬럼명(Column names) : 해당 열의 데이터들을 대표하는 이름
* 인덱스(Index labels) : 해당 행을 고유하게 식별하는 개체

#### 2. Series

* series 생성

```
# 딕셔너리 형태로 Series 생성
population = pd.Series({'서울':9904312, '부산':3448737,'인천':2890451, '대구':2466052})
population
```

```
# 리스트를 이용하여 Series 생성
population = pd.Series([9904312, 3448737, 2890451, 2466052],
                       index = ['서울', '부산', '인천', '대구'])
population
```

```
# Series명 설정
population.name = '인구'
# index명 설정
population.index.name = '도시'

population
```

* Series 데이터 갱신, 추가, 삭제

```
# 데이터 갱신
population['부산'] = 3500000

# 데이터 추가
population['광주'] = 2400000

# 데이터 삭제
del population['서울']

population
```

* Series 속성 확인하기


```
# Series의 데이터 값 확인
population.values
```

```
# series 인덱스 확인
population.index
```

```
# Series 타입 확인
population.dtype
```

#### 3. DataFrame

* DataFrame 생성

```
# 딕셔너리 형태로 DF 생성
data = {'2015':[9904312,3448737,2890451,2466052],
        '2010':[9631482,3393191,2632035,2471774]}

population_df = pd.DataFrame(data)
population_df.index = ['서울','부산','인천','대구']
population_df
```

```
# 리스트를 이용하여 DF 생성
# data 값
data = [[9904312,3448737,2890451,2466052],
        [9631482,3393191,2632035,2471774]]
# index 값
year = ['2015','2010']
# columns 명
city = ['서울','부산','인천','대구']

population_df = pd.DataFrame(data, index = year, columns = city)
population_df
```

```
# DF의 행과 열 체인지
population_df = population_df.T
population_df
```

* DF 데이터 추가, 삭제

```
# 데이터 추가
population_df['2005'] = [9762546, 3512547, 2517680, 2456016]
population_df['2000'] = [9762546, 3512547, 2517680, 2456016]

population_df
```

```
# 데이터 삭제(del)
del population_df['2005']

# 데이터 삭제(drop)
# axis = 0 : 행방향, 1 : 열 방향
population_df = population_df.drop('2000', axis=1)

population_df
```

* DF 속성 확인하기

```
# DF의 형태와 크기 확인
population_df.shape
```

```
# DF의 데이터 값을 확인
population_df.values
```

```
# DF의 columns 확인
population_df.columns
```

```
# DF의 index 확인
population_df.index
```

#### 4.인덱싱, 슬라이싱

##### Series

```
score = pd.Series({'java':70, 'python':95, 'html/css':80, 'ml':100})
score 
```

```
# 인덱스 번호로 접근
print(score[1])
print(score[3])
# 인덱스 명으로 접근
print(score['html/css'])
```

```
# 인덱스 명 자체로 접근
score.index[1]
```

```
# 인덱스 번호(마지막 값 포함x)로 범위를 지정하여 인덱싱, 슬라이싱
print(score[1:])

print('-'*25)

# 인덱스 명(마지막 값 포함)으로 범위를 지정하여 인덱싱, 슬라이싱
print(score['python':'ml'])
```

* Boolean 인덱싱

```
print(score >= 85)

print('-'*25)

print(score[score >= 85])
```

##### DataFrame

```
stu_score = pd.DataFrame({'java':[95,85,70],'python':[100,95,90],'ml':[85,85,100],'html/css':[70,75,80]}, 
                         index = ['mj','sy','sb'])
stu_score
```

```
stu_score[0:]
```

```
stu_score['mj':'sy']
```

* Boolean 인덱싱

```
stu_score[stu_score['ml'] > 90]
```

```
stu_score[stu_score['python']>90]
```

* loc[행 인덱스, 열 인덱스] : 실제 인덱스를 사용하여 행을 가지고 올 때 사용
* iloc[array행 인덱스, array열 인덱스] : numpy의 array인덱싱 방식으로 행을 가지고 올때 사용

```
stu_score.loc['mj']
```

```
stu_score.loc['mj','java']
```

```
stu_score.iloc[1]
```

```
stu_score.iloc[1,2]
```
