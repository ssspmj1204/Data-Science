#### 3. 영화 아이디 1번인 영화의 평점 평균 구하기 

```python
# 영화 아이디
item_id = data[:,1]
item_id
```

```python
item_id == 1
```

```python
# 영화 아이디가 1번인 데이터 
data[item_id == 1]
```

```python
# 영화 아이디가 1번인 데이터의 평점
data[item_id == 1][:,2]
```

```python
# 영화 아이디가 1번인 데이터의 평점 평균
data[item_id == 1][:,2].mean()
```

#### 4.  각 영화의 평점 평균 구하여 list에 저장

```python
item_id
item_id_un = np.unique(item_id)
```

```python
item_id_un
```

```python
len(item_id_un)
```

```python
result = []
for i in item_id_un:
    result.append([i,data[item_id == i][:,2].mean()])
```

```python
print(result, end = ' ')
```

```python
len(result)
```

#### 5. 각 영화의 평점 평균을 담은 list 중에서 평점 평균이 4이상인 영화 아이디 출력 

```python
count = 0

print("평점 평균이 4이상인 영화 아이디: ")
for i in range(len(result)):
    if(result[i][1] >= 4):
        print(result[i][0], end = ' ')
        count+=1
        
print("\n\n")
print("평점 평균이 4이상인 영화의 수: ",count)
```
