```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime_data = pd.read_csv("경찰청_연도별 사이버범죄통계.csv", encoding = "euc-kr", index_col = "구분")
crime_data = crime_data.sort_index()
crime_data
```

```python
occur = crime_data[crime_data['구분.1'] == "발생건수"]
occur = occur.T
occur.drop('구분.1', inplace = True)
occur.name = '발생'
occur
```

```python
arrest = crime_data[crime_data['구분.1'] == "검거건수"]
arrest = arrest.T
arrest.drop('구분.1', inplace = True)
arrest.name = '검거'
arrest
```

```python
from matplotlib import font_manager,rc
rc('font', family = 'NanumGothic')
```

```python
x = occur.iloc[0]
x_name = occur.iloc[0].name
```

```python
plt.plot(x, label = occur.name)
plt.xlabel("발생연도")
plt.ylabel("발생건수")
plt.title(x_name)
plt.legend()
plt.show()
```

```python
for i in range(33): 
    occur_num = occur.iloc[i]
    arrest_num = arrest.iloc[i]
    crime_name = occur.iloc[i].name
    plt.plot(occur_num, label = occur.name, marker = 'o', color = 'b')
    plt.plot(arrest_num, label = arrest.name, marker = 'o', color = 'r')
    plt.xlabel("발생연도")
    plt.ylabel("건수")
    plt.title(crime_name)
    plt.legend()
    plt.show()
```
